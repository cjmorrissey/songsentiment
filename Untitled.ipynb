They look to be just place holders, however after doing some poking around not shown here, there are various other things in the text that appear. Things like "[Chorus]" or "[Chorus: x2]" so we will remove them -- it turned out while going through this project words like "verse" were appear in the wordclouds to come, so I chose to remove them.

However it is worth noting, some songs do make use of these words -- for example Justin Timberlake's 'Sexyback' has the line "Because you're burning up, I got to get it fast (yeah, take it to the bridge)". So we should try to be careful.

```python
drop_list = ['\[','\]','x2', 'Chorus:', 'Chorus', 'Verse', 'Hook', 'Bridge']
for item in drop_list:
    df['text'] = df['text'].replace(item, '', regex = True)
```

Some basic stats

```python
# Get the how many words each 'Artist' has?
def get_artist_rep(artist):
    counts = {}
    for x in artist:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
def max_artist(artist):
    max_ind = -1
    for k in artist:
        if(artist[k] > max_ind):
            max_ind = artist[k]
    return k, max_ind
```

```python
count_artist = get_artist_rep(df['artist'])
artist, max_ind = max_artist(count_artist)
print("'"+ artist +"'performed "+ str(max_ind) + " songs.")
```

```python
def definition_score(df, score_word_dict):
    score = 0
    words = df.lower().split(" ")
    for word in words:
        if(word in score_word_dict):
            score += int(score_word_dict[word])
            
    return score
```

```python
scores=[]
for d in df['text']:
    score = definition_score(d, score_word_dict)
    #print(score)
    scores.append(score)
```

```python
df['definition_feeling_score'] = scores
```
